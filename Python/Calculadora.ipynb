# Projeto Calculadora em Python

Objetivo de criar uma calculadora em Python, utilizando funções e condicionais.

```python
# declarando as funções da calculadora

def somar(x, y):
  return x + y

def subtrair(x, y):
  return x - y

def multiplicar(x, y):
  return x * y

def dividir(x, y):
  return x / y
```

```python
# mostrando as opções de operações para o usuário

print('=============== Calculadora Python ===============')
print('\n1 - Somar\n2 - Subtrair\n3 - Multiplicar\n4 - Dividir')

# solicitando uma opção ao usuário

operacao = int(input('\nDigite uma opção da operação a ser feita: '))
print('\n')

# condição para que a operação seja executada

if operacao > 0 and operacao <= 4:

# solicitando os números ao usuário

  num1 = float(input('Digite o primeiro número: '))
  num2 = float(input('Digite o segundo número: '))

# condicional para cada opção escolhida pelo usuário

  if operacao == 1:
    print(num1, '+', num2, '=', somar(num1, num2))
    print('\n')
  elif operacao == 2:
    print(num1, '-', num2, '=', subtrair(num1, num2))
    print('\n')
  elif operacao == 3:
    print(num1, 'x', num2, '=', multiplicar(num1, num2))
    print('\n')
  elif operacao == 4:
    print(num1, ':', num2, '=', dividir(num1, num2))
    print('\n')
else:
  print('Opção inválida! Tente novamente.')
  print('\n')
print(51 * '=')
```
