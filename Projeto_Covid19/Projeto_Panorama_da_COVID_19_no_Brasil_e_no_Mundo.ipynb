# Panorama do COVID-19 no Brasil e no Mundo

COVID-19 é uma doença infecciosa causada pelo vírus SARS-Cov-2, identificada pela primeira vez em dezembro de 2019 na cidade de Wuhan, na China, com um surto de pneumonia por causa desconhecida. Houve relatos que as pessoas infectadas tiveram contatos com animais em um mercado de frutos do mar e animais vivos na cidade, mas não foi confirmado que o vírus tenha sido transmitido por estes animais.

Em Janeiro de 2020, o vírus se espalhou rapidamente em outras partes da China e depois para outros países, quando em Março de 2020 foi decretada pela OMS (Organização Mundial da Saúde) como pandemia global.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa, mas podem incluir febre, tosse seca, fadiga, dor de cabeça, perda de olfato ou paladar, dor de garganta, dores musculares e dificuldades na respiração. Em casos mais graves, a doença pode levar a pneumonia, falência de orgãos e até a morte.

Desde o início da pandemia, a COVID-19 afetou milhões de pessoas em todo o mundo e teve um impacto significativo na saúde pública, economia e na sociedade como um todo.

Após alguns meses de pesquisas, as vacinas foram liberadas no mês Dezembro de 2020, tornando-se uma medida muito importante para controlar a proliferação da doença.

Mesmo após a liberação das vacinas, notou-se um aumento de casos principalmente em épocas de festas e datas comemorativas, fazendo com que o vírus sofresse mutações gerando novas variantes e consequentemente novas doses de vacinas.

Neste **projeto** farei a análise de uma maneira geral sobre a COVID-19 no Brasil e no Mundo, para assim conseguir extrair alguns insights importantes dos dados.

### 2.1.Quais as regiões no mundo possuem mais casos e mortes de COVID-19?

Nesta fase do projeto, vamos ter uma visão geral sobre qual região, geograficamente falando, foi mais afetada com a pandemia causada pela doença COVID-19.

```
## verificando os casos de COVID-19 atualmente por localidade geográfica em ordem decresente
data_atual_casos = df_covid.loc[df_covid['date'] == '2023-03-03'].sort_values(by='total_cases', ascending=False)[1:7]
data_atual_casos
```

Sobre as regiões pelo mundo que mais tem casos registrados da COVID-19:

* Na primeira posição aparece os países de alta-renda com 431.389.174 casos, pode ser por serem países mais populosos
* A Europa é a região com mais casos registrados, totalizando 248.178.623 casos
* A Ásia é a segunda região com mais casos de COVID-19, totalizando 210.624.451 casos
* A América do Norte, formada por apenas 3 países, Estados Unidos, Canadá e México, vem em terceiro lugar com 122.518.346 casos


```
## verificando as mortes de COVID-19 atualmente por localidade geográfica em ordem decresente
data_atual_casos = df_covid.loc[df_covid['date'] == '2023-03-03'].sort_values(by='total_deaths', ascending=False)[1:6]
data_atual_casos
```

Falando de mortes por COVID-19 no mundo:

* A Ásias fica em segundo lugar no número de mortes, 1.623.045
* A América do Norte, vem em terceiro no ranking, com 1.586.062 mortes, lembrando que que está região é composta por apenas 3 países, Estados Unidos, Canadá e México

### 2.2.Considerando a data atual, quais os países com mais casos e mortes?

Veremos então como estão os casos de mortes por COVID-19 considerando a data mais atual deste dataset, que é 03.03.2023, lembrando que este dataset pode ter sua data atual alterada para mais recente no arquivo oficial.

```
# verificando os casos de COVID-19 atualmente por países em ordem decresente
data_atual_casos = df_covid.loc[df_covid['date'] == '2023-03-03'].sort_values(by='total_cases', ascending=False)[7:14]
data_atual_casos
```

Quando tratamos de casos de COVID-19 registrados até o momento por país, podemos afirmar que:

* Os Estados Unidos lidera o ranking com 103.643.742 casos
* O Brasil é o quinto país com mais casos registrados, 37.081.209 casos
* Qeem aparece em segundo lugar com maior número de casos é a Índia, 44.688.722 casos

```
# verificando os casos de mortes atualmente por países em ordem decrescente
data_atual_mortes = df_covid.loc[df_covid['date'] == '2023-03-03'].sort_values(by='total_deaths', ascending=False)[9:14]
data_atual_mortes
```

Se tratando da quantidades de mortes por país, até a data de 03 de Março de 2023:

* Os Estados Unidos aparece em primeiro lugar no ranking com 1.122.157 mortes
* O Brasil vem em segundo lugar com 699.276 mortes
* Em terceiro lugar com maior número de mortes está a Índia com 530.775 mortes

Dado importante é que, a Índia é o segundo país com maior número de casos registrados, porém, em número de mortes fica atrás do Brasil que é o quinto país em número de casos.

Ou seja, morre mais pessoas no Brasil com menor quantidade de casos do que na Índia.

Para que a gente possa visualizar melhor os resultados de mortes, vou plotar um gráfico de barras.

```
# visualizando o total de mortes através de um gráfico de barras
data_atual_mortes10 = df_covid.loc[df_covid['date'] == '2023-03-03', ['location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)[9:15]

data_atual_mortes10.plot(x='location', y='total_deaths', kind='bar', figsize=(10, 5), color='orange')
plt.title('Total de mortes - COVID-19')
plt.xlabel('Países')
plt.ylabel('Mortes')

plt.show();
```

### 2.3.Quantas pessoas no Mundo receberam no mínimo a primeira dose da vacina?

Agora vamos analisar quantas pessoas no mundo já foram vacinadas com pelo menos uma dose de vacina para COVID-19.

```
#visualzando os valores máximos da variável World 
df_covid.loc[df_covid['location'] == 'World'].max()
```

```
#pessoas vacinadas no mundo em relação a população total
populacao_mundo = 7975105024
pessoas_vacinadas_mundo = 5557339852
total_vacinados_mundo = (pessoas_vacinadas_mundo / populacao_mundo) * 100

print('{:.2f} das pessoas no Mundo receberam no mínimo a primeira dose de vacina.'.format(total_vacinados_mundo))
```

Após analisar os dados acima, chego a conclusão que aproximadamente 70% da população Mundial tomou pelo menos uma dose da vacina contra a COVID-19.

### 2.4.Evolução do vírus pelo Mundo

Uma das maiores áreas que se preocupa com a proliferação do vírus pelo mundo é a área da saúde. Mas não somente a área da saúde é afetada quando isso ocorre, toda área financeira e comercial de um país acaba entrando em crise com uma declaração de pandemia.


```
# plotando um gráfio de linhas 
df_covid.loc[df_covid['location'] == 'World', 'total_deaths'].reset_index(drop=True).plot(figsize=(10, 5));
```

Porém conforme as doses de vacinas foram sendo aplicadas, houve um maior controle nos casos de mortes. Não deixamos de pegar o vírus, apenas não estamos mais tendo complicações de saúde.

### 4.Evolução do vírus no Brasil

Nesta última etapa do projeto, vamos analisar através dos gráficos, como foi a evolução do vírus no Brasil para os casos de infecção e casos de mortes pela COVID-19.

```
# plotando os gráficos de casos e mortes em escalas linear e logarítmica
a = list(range(1102))
b = df_brasil['total_cases']
c = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize=(10, 5))
ax1.set_title('Casos de COVID-19 no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(a, b)

ax2.set_title('Casos de COVID-19 no Brasil (escala logarítmica')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(a, b)

ax3.set_title('Casos de mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(a, c)

ax4.set_title('Casos de mortes no Brasil (escala logarítmica')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(a, c)

plt.tight_layout()
```

Podemos concluir analisando os gráficos acima, que para os casos de infecção da doença os número foram subindo aos poucos até atingirem o auge, com relação as mortes elas demoraram um pouco para atingir o pico, e quando atingiu se manteve em alta.
